# 1. Line Plot

```python
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5,6,7,8,9]
temperature = [32,34.2,43,54,34,32,34.2,43,54]
```

```python
plt.plot(days,temperature)
# plt.axis([0,10,0,60])
plt.title("Bihar Temprature")
plt.xlabel("days")
plt.ylabel("temprature")
plt.show()
```

```python
plt.plot(days,temperature)
plt.axis([0,10,0,60])
plt.title("Bihar Temprature")
plt.xlabel("days")
plt.ylabel("temprature")
plt.show()
```

### Fore more informative

```python
plt.plot(days,temperature, color = "r", marker = "o", linestyle = "--", linewidth = 3, markersize = 9) #this is very lendy
plt.axis([0,10,0,60])
plt.title("Bihar Temprature")
plt.xlabel("days")
plt.ylabel("temprature")
plt.show()
```

```python
# in Sort (color,marker,linestyle)
```

```python
plt.plot(days,temperature, "o--", linewidth = 3, markersize = 9) # label="temp line" also use here
plt.axis([0,10,0,60])
plt.title("Bihar Temprature", fontsize = 22)
plt.xlabel("days", fontsize = 22)
plt.ylabel("temprature", fontsize = 22)
plt.legend(["Temp line"], loc = 4)
plt.show()
```

### Style in matplotlib

```python

from matplotlib import style

style.use("ggplot")

plt.plot(days,temperature, "m--", linewidth = 3, markersize = 9) # label="temp line" also use here
plt.axis([0,10,0,60])

plt.title("Bihar Temprature", fontsize = 22)

plt.xlabel("days", fontsize = 22)
plt.ylabel("temprature", fontsize = 22)

plt.legend(["Temp line"], loc = 4)

plt.grid(color='r', linestyle='-', linewidth=2)
plt.show()
```

### add Another curve

```python
from matplotlib import style
day = [1,2,3,4,5,6,7,8,9]
temp = [41,24.2,23,44,34,12,34.2,43,54]
style.use("ggplot")

plt.plot(days,temperature, "m--", linewidth = 3, markersize = 9, label="Bihar") # label="temp line" also use here
plt.plot(day,temp, "r--", linewidth = 3, markersize = 9, label="Delhi")
plt.axis([0,10,0,60])
plt.title("Bihar & Delhi Temprature", fontsize = 22)
plt.xlabel("days", fontsize = 22)
plt.ylabel("temprature", fontsize = 22)
plt.legend( loc = 4)

plt.grid(color='r', linestyle='-', linewidth=2)
plt.show()
```
