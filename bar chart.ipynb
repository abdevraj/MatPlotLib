```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ['python','C','R','AI','ML']
class1_std = [33,23,34,19,40]
class2_std = [43,13,34,29,45]
class3_std = [63,23,54,29,30]
```

```python
plt.bar(classes,class1_std)
```

```python
plt.barh(classes,class1_std)
```

```python
plt.bar(classes,class1_std, width=0.2,align="edge",color='y',edgecolor='m', linewidth=3, alpha=0.5)
```

```python
plt.bar(classes,class1_std, width=0.2,align="edge",color='y',edgecolor='m', linewidth=3, alpha=0.5,
        linestyle='--',label="abc classes" )
```

```python
# plt.figure(figsize=(16,9))
plt.bar(classes,class1_std, width=0.2,align="edge",color='y',edgecolor='m', linewidth=3, alpha=0.5,
        linestyle='--',label="abc classes" ) # visible=False
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1_std, width=0.2,align="edge",color='y',edgecolor='m', linewidth=3, alpha=0.5,
        linestyle='--',label="abc classes" )
```

```python
plt.figure(figsize=(16,9))
plt.barh(classes,class1_std, align="edge",color='y',edgecolor='m', linewidth=3, alpha=0.5,
        linestyle='--',label="abc classes" ) # in barh we cant use width 

```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_std,width=0.2, align="edge",color='y',edgecolor='m', linewidth=3, alpha=0.5,
        linestyle='--',label="abc classes" ) 

plt.title("bar chaet of ABC class", fontsize=20)
plt.xlabel("classes",fontsize=15)
plt.ylabel("no opf students",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_std,width=0.2, align="edge",color='y',edgecolor='m', linewidth=3, alpha=0.5,
        linestyle='--',label="class1" ) 
plt.bar(classes,class2_std,width=0.2, align="edge",color='g',edgecolor='b', linewidth=3, alpha=0.5,
        linestyle='--',label=" class2" )

plt.title("bar chaet of ABC class", fontsize=20)
plt.xlabel("classes",fontsize=15)
plt.ylabel("no opf students",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,9))

classes_index=np.arange(len(classes))
width=0.2

plt.bar(classes_index,class1_std,width, color='y',label="class1" ) 

plt.bar(classes_index + width,class2_std,width,color='g',label=" class2" )

plt.bar(classes_index + width + width,class3_std,width,color='k',label=" class3" )

plt.xticks(classes_index, classes, rotation = 20)
plt.title("bar chaet of ABC class", fontsize=20)
plt.xlabel("classes",fontsize=15)
plt.ylabel("no opf students",fontsize=15)
plt.legend()
plt.show()
```
