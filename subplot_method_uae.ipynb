```python
 plt.figure(figsize=(16,9))

# --------------------------------------------------start
plt.subplot(3,2,1)
#----------------------------Line plot
day = [1,2,3,4,5,6,7,8,9]
temp = [41,24.2,23,44,34,12,34.2,43,54]
temprature = [47,24.2,43,24,34,32,34.2,23,24]

plt.plot(day,temp, "m--", linewidth = 3, markersize = 9, label="Bihar") # label="temp line" also use here
plt.plot(day,temprature, "r--", linewidth = 3, markersize = 9, label="Delhi")
plt.axis([0,10,0,60])
plt.title("Bihar & Delhi Temprature", fontsize = 22)
plt.xlabel("days", fontsize = 15)
plt.ylabel("temprature", fontsize = 18)
plt.legend( loc = 4)

plt.grid(color='r', linestyle='-', linewidth=2)
# --------------------------------------------------end

# --------------------------------------------------start
plt.subplot(3,2,2)
# --------------------------------------------------Histagram plot
ml_student_age=np.random.randint(18,45,(100))
py_student_age=np.random.randint(15,40,(100))

bins = [15, 20, 25, 30, 35, 40, 45]

plt.hist([ml_student_age, py_student_age], bins, rwidth=0.8, orientation='vertical', color=['m','y'], label=['ML Student','ML Student'])


plt.title("student age of ml & py list")
plt.xlabel("Student age groups")
plt.ylabel("No of students age")
plt.legend()
plt.grid(color="k", linestyle="--", linewidth=8)
# --------------------------------------------------end

# --------------------------------------------------start
plt.subplot(3,2,3)
# --------------------------------------------------Bar chart

classes = ['python','C','R','AI','ML']
class1_std = [33,23,34,19,40]
class2_std = [43,13,34,29,45]
class3_std = [63,23,54,29,30]

classes_index=np.arange(len(classes))
width=0.2

plt.bar(classes_index,class1_std,width, color='y',label="class1" ) 

plt.bar(classes_index + width,class2_std,width,color='g',label=" class2" )

plt.bar(classes_index + width + width,class3_std,width,color='k',label=" class3" )

plt.xticks(classes_index, classes, rotation = 20)
plt.title("bar chaet of ABC class", fontsize=20)
plt.xlabel("classes",fontsize=15)
plt.ylabel("no opf students",fontsize=15)
plt.legend()

# --------------------------------------------------end
plt.subplot(3,2,4)
#-----------------------------------start
df_google_play_store_apps = pd.read_csv("googleplaystore.csv", nrows = 1000)

x = df_google_play_store_apps["Rating"]
y = df_google_play_store_apps["Reviews"]

plt.scatter(x,y, c='r', marker='*', s=100, alpha=0.6, linewidth=9, edgecolor='y')

plt.scatter(x,df_google_play_store_apps["Installs"], c='g', marker='*', s=100, alpha=0.6, linewidth=9, edgecolor='m')

plt.title("Google play store scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
#------------------------------------end

#------------------------------------start
plt.subplot(3,2,5)
#--------------------------------------------pie plot
classes = ['Python', 'R', "Machine Learning", 'Artificial Intelligence', 'Data Science']
class1_student = [45,15,35,25,30]

explode = [0.03,0,0.2,0,0]  # VALUE BETWEEN O - 1
colors = ['c','b','r','y','g']  #change colors
#  autopct ="%0.2f%%" --->  use for show percentage
#  radius = 1.4  ---->  use for icrease pic chart
textprops ={"fontsize":15}
wedgeprops = {"linewidth": 2, "width":1, "edgecolor":"k"} 

plt.pie(class1_student, labels = classes, explode = explode, colors = colors, autopct ="%0.2f%%", shadow = True, radius = 1.4,
       startangle = 90, textprops = textprops, wedgeprops = wedgeprops)
plt.legend()
#--------------------------------------------end
plt.subplot(3,2,6, projection="polar", facecolor='k', frameon=True)

plt.savefig("subplot")
plt.show()
```
