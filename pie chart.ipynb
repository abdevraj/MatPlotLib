```python
import matplotlib.pyplot as plt
```

```python
classes = ['Python', 'R', "Machine Learning", 'Artificial Intelligence', 'Data Science']
class1_student = [45,15,35,25,30]
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(class1_student, labels = classes)
plt.show()
```

```python
explode = [0.03,0,0.2,0,0]  # VALUE BETWEEN O - 1
colors = ['c','b','r','y','g']  #change colors
#  autopct ="%0.2f%%" --->  use for show percentage
#  radius = 1.4  ---->  use for icrease pic chart
textprops ={"fontsize":15}
wedgeprops = {"linewidth": 2, "width":1, "edgecolor":"k"} 

plt.pie(class1_student, labels = classes, explode = explode, colors = colors, autopct ="%0.2f%%", shadow = True, radius = 1.4,
       startangle = 90, textprops = textprops, wedgeprops = wedgeprops)
plt.legend()
plt.show()
```

```python
classes = ['Python', 'R', "Machine Learning", 'Artificial Intelligence', 'Data Science']
class1_student = [45,15,35,25,30]

explode = [0.03,0,0.2,0,0]  # VALUE BETWEEN O - 1
colors = ['c','b','r','y','g']  #change colors
#  autopct ="%0.2f%%" --->  use for show percentage
#  radius = 1.4  ---->  use for icrease pic chart
textprops ={"fontsize":15}
wedgeprops = {"linewidth": 2, "width":1, "edgecolor":"k"} 

plt.pie(class1_student, labels = classes, explode = explode, colors = colors, autopct ="%0.2f%%", shadow = True, radius = 1.4,
       startangle = 90, textprops = textprops, wedgeprops = wedgeprops)
plt.legend()
```
