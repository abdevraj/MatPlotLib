```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ml_student_age=np.random.randint(18,45,(100))
py_student_age=np.random.randint(15,40,(100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)

plt.title("student age of ml & py list")
plt.xlabel("Student age groups")
plt.ylabel("No of students age")
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]

plt.figure(figsize=(16,9))

plt.hist([ml_student_age, py_student_age], bins, rwidth=0.8, orientation='vertical', color=['m','y'], label=['ML Student','ML Student'])

# plt.hist(py_student_age, bins, rwidth=0.8, orientation='vertical', color='r', label='PY Student')

plt.title("student age of ml & py list")
plt.xlabel("Student age groups")
plt.ylabel("No of students age")
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(16,9))

plt.hist([ml_student_age, py_student_age], bins, rwidth=0.8, orientation='vertical', color=['m','y'], label=['ML Student','ML Student'])

# plt.hist(py_student_age, bins, rwidth=0.8, orientation='vertical', color='r', label='PY Student')

plt.title("student age of ml & py list")
plt.xlabel("Student age groups")
plt.ylabel("No of students age")
plt.legend()
plt.grid(color="k", linestyle="--", linewidth=8)
plt.show()
```
